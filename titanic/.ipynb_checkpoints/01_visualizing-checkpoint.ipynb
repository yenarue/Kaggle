```python
import pandas as pd
pd.options.mode.chained_assignment = None 

## 파일 읽기
test = pd.read_csv('./test.csv')
train = pd.read_csv('./train.csv')
```

## 시각화
타이타닉 트레이닝 데이터의 각 항목별로 데이터 시각화를 해보자.

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set() # setting seaborn default for plots
```

```python
def bar_chart(feature):
    survived = train[train['Survived'] == 1][feature].value_counts()
    dead = train[train['Survived'] == 0][feature].value_counts()
    df = pd.DataFrame([survived, dead])
    df.index = ['Survived', 'Dead']
    df.plot(kind = 'bar', stacked = True, figsize = (10, 5))
```

```python
bar_chart('Sex')
```

여자들이 남자들보다 많이 살아남았구나.
=> **[가설] 여자는 남자보다 살 확률이 높다.**

```python
bar_chart('Pclass')
```

객실 클래스 차이가 죽음에 영향을 미칠 수 있겠구나
**=> [가설] 객실(좌석)의 클래스가 좋을수록 살 확률이 높다.**

```python
bar_chart('SibSp')
```

혼자 탄 사람들은 죽은 사람이 더 많구나 => **[가설] 가족이 있는 경우에는 살 확률이 더 높다.**

```python
bar_chart('Parch')
```

혼자 탄 사람들은 죽은 사람이 더 많구나
=> **[가설] 부모, 아이가 있는 경우에는 살 확률이 더 높다.**

```python
bar_chart('Embarked')
```

S,Q 선착장에서 탄 사람들은 더 많이 죽었네.. C 선착장에서 탄 사람은 생존자와 사망자가 반반 같다.

### Name (이름) -> Title 로 가공
Mr, Mrs, Miss 등의 타이틀 정보를 학습에 사용해보자.
생존자에 여성이 많았다는 점에 착안하여, **아이가 있거나 남편이 있는 여자가 살아남을 확률이 높았다**고 가정해보자.

일단 `Name` 필드를 가공하여 `Title` 필드를 새로 만들어주자.

```python
train_test_data = [train, test]

for dataset in train_test_data:
    dataset['Title'] = dataset['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
```

```python
train['Title'].value_counts()
```

```python
train.head()
```

`Title` 정보를 데이터화 시키기 편하도록 숫자로 매핑해준다.
* 0 : 남성 (Mr)
* 1 : 여성 (Miss)
* 2 : 기혼 여성 (Mrs)

```python
title_mapping = {"Mr": 0, "Miss": 1, "Mrs": 2, 
                 "Master": 3, "Dr": 3, "Rev": 3, "Col": 3, "Major": 3, "Mlle": 3,"Countess": 3,
                 "Ms": 3, "Lady": 3, "Jonkheer": 3, "Don": 3, "Dona" : 3, "Mme": 3,"Capt": 3,"Sir": 3 }

for dataset in train_test_data:
    dataset['Title'] = dataset["Title"].map(title_mapping)
```

```python
bar_chart('Title')
```

```python
train.drop('Name', axis=1, inplace=True)
test.drop('Name', axis=1, inplace=True)
```

```python
test.head()
```

### Sex (성별)

```python
sex_mapping = {"male" : 0, "female" : 1}
for dataset in train_test_data:
    dataset['Sex'] = dataset['Sex'].map(sex_mapping)
```

#### Binning
나이와 같이 너무나도 시퀀셜한 데이터인 경우에, 카테고라이징 하는 기법

* child : 0
* young : 1
* adult : 2
* mid-age : 3
* senior : 4

```python
for dataset in train_test_data:
    dataset.loc[dataset["Age"] <= 16, "Age"] = 0,
    dataset.loc[(dataset["Age"] > 16) & (dataset["Age"] <= 26), "Age"] = 1,
    dataset.loc[(dataset["Age"] > 26) & (dataset["Age"] <= 36), "Age"] = 2,
    dataset.loc[(dataset["Age"] > 36) & (dataset["Age"] <= 62), "Age"] = 3,
    dataset.loc[dataset["Age"] > 62, "Age"] = 4
```

```python
train.head()
```

```python
bar_chart('Age')
```

어린아이들은 조금 더 많이 살았다. 나이 드신 분들은 거의 죽었다.

### Embarked & Pclass

`Embarked` 필드는 약간의 누락이 존재한다 (889/891)

#### 누락된 값 채우기
[가설] 부자 도시의 선착장에서 탄 사람들은 더 등급이 높은 클래스를 탔을 것이다.

```python
Pclass1 = train[train['Pclass'] == 1]['Embarked'].value_counts()
Pclass2 = train[train['Pclass'] == 2]['Embarked'].value_counts()
Pclass3 = train[train['Pclass'] == 3]['Embarked'].value_counts()
df = pd.DataFrame([Pclass1, Pclass2, Pclass3])
df.index = ['1st class', '2nd class', '3rd class']
df.plot(kind='bar', stacked=True, figsize=(10, 5))
```

Q 선착장에서 탄 사람들은 1등급에 거의 없다.
S 선착장에서 탄 사람들이 매우 많다. (과반수 이상?!)

누락된 필드는 총 2개 뿐이다.
=> 그냥 누락된 `Embarked` 필드에는 `S` 라고 넣어도 무방할 듯 하다.

```python
for dataset in train_test_data:
    dataset['Embarked'] = dataset['Embarked'].fillna('S')
```

```python
train.head()
```

`Embarked` 정보도 숫자로 치환해준다!

```python
embarked_mapping = {"S" : 0, "C" : 1, "Q" : 2}
for dataset in train_test_data:
    dataset["Embarked"] = dataset["Embarked"].map(embarked_mapping)
```
